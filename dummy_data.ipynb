```python
#Fix the professional certification should be C COOK,B COOK,A COOK,SOUS CHEF,CHEF
with open("Chefs_dummy_data.json") as f1, open("DML_SCRIPT.sql", "w") as f2:
    data = json.load(f1)
    prof_certifications = {0:"None" , 1:"C COOK", 2:"B COOK", 3:"A COOK", 4:"SOUS CHEF", 5:"CHEF"}
    f2.write("USE CookingContest;\n")
    f2.write("\n")
    def generate_prof_certification(work_experience):
        work_experience = work_experience % 10
        if work_experience < 2:
            return random.randint(0,1)
        elif work_experience < 4:
            return random.randint(1,2)
        elif work_experience < 6:
            return random.randint(2,3)
        elif work_experience < 8:
            return random.randint(3,4)
        else:
            return random.randint(4,5)
        return 0
    for chef in data:
        prof_cert = generate_prof_certification(chef['work_experience'])
        surname = chef['surname'].replace('\'','')
        f2.write(f"INSERT INTO Chefs (name, surname , phone_number, birth_date, work_experience, prof_certification) VALUES ('{chef['name']}', '{surname}', '{chef['phone_number']}', '{chef['birth_date']}', '{chef['work_experience'] % 10}', {prof_cert});\n")
```

```python
recipes = []
ingredients = []
tags = []
meals = []
themes = []
utensils = []
cuisines = []
quantities = []
group_map = {
    "Vegetables": 1,
    "Fruit": 2,
    "Wheat and grains": 3,
    "Dairy products": 4,
    "Legumes": 5,
    "Meat": 6,
    "Beef" : 6,
    "Chicken": 7,
    "Eggs": 8,
    "Seafood": 9,
    "Oils": 10,
    "Nuts" : 10,
    "N/A": 11
}
with open("Recipes.json", "r", encoding="utf-8") as f1 , open("DML_SCRIPT.sql", "a", encoding="utf-8") as f2:
    recipes_info = json.load(f1)
    for (index, recipe) in enumerate(recipes_info):
        ingredients_info = recipe['ingredients']#
        main_ingredient = recipe['main_ingredient']
        for ingredient_info in ingredients_info:
            main_ing = False
            food_group_index = group_map[ingredient_info[1]]
            calories_per_100g = int(ingredient_info[2])
            ingredient_quantity = ingredient_info[3]
            numeric_quantity = random.randint(200,500)
            name = ingredient_info[0].replace('\'','')
            #Compare the name of the ingredient to the main ingredient loosely
            if name.lower() == main_ingredient.lower():
                main_ing = True
            if name not in ingredients:
                ingredients.append(name)
                f2.write(f"INSERT INTO Ingredients (food_group_id, name, calories_per_100_units) VALUES ({food_group_index}, '{name}', {calories_per_100g});\n")
            f2.write(f"INSERT INTO requires (recipe_id, ingredient_id, quantity, num_units, main_ingredient) VALUES ({index + 1}, {ingredients.index(name) + 1}, '{ingredient_quantity}', {numeric_quantity}, {int(main_ing)});\n")
        if recipe["theme"] not in themes:
            themes.append(recipe["theme"])
        theme = recipe["theme"]
        f2.write(f"INSERT INTO Themes (name,descr) VALUES ('{theme}', 'N/A');\n")
        f2.write(f"INSERT INTO recipe_theme (recipe_id, theme_id) VALUES ({index + 1}, {themes.index(theme) + 1});\n")
        for (step_index, step) in enumerate(recipe['steps']):
            step = step.replace('\'','')
            f2.write(f"INSERT INTO Steps (recipe_id, step_number, step_desc) VALUES ({index + 1}, {step_index + 1}, '{step}');\n")
        quantity_of_servings = recipe['quantity']
        f2.write(f"INSERT INTO Quantities (recipe_id, quantity) VALUES ({index + 1}, {quantity_of_servings});\n")
        for tag in recipe['tags']:
            if tag not in tags:
                tags.append(tag)
                f2.write(f"INSERT INTO Tags (tag_name) VALUES ('{tag}');\n")
            f2.write(f"INSERT INTO recipe_tags (recipe_id, tag_id) VALUES ({index + 1}, {tags.index(tag) + 1});\n")
        for meal in recipe['meal_types']:
            if meal not in meals:
                meals.append(meal)
                f2.write(f"INSERT INTO meals (name) VALUES ('{meal}');\n")
            f2.write(f"INSERT INTO recipe_meal (recipe_id, meal_id) VALUES ({index + 1}, {meals.index(meal) + 1});\n")
        for utensil in recipe['kitchenware']:
            if utensil[0] not in utensils:
                utensil_name = utensil[0].replace('\'','')
                utensils.append(utensil_name)
                utensil_desc = utensil[1].replace('\'','')
                f2.write(f"INSERT INTO Kitchenware (name,usage_desc) VALUES ('{utensil_name}', '{utensil_desc}');\n")
            f2.write(f"INSERT INTO kitchenware_for_recipe (recipe_id, kitchenware_id) VALUES ({index + 1}, {utensils.index(utensil_name) + 1});\n")
        if recipe['cuisine'] not in cuisines:
            cuisines.append(recipe['cuisine'])
            f2.write(f"INSERT INTO Cuisines (country_name) VALUES ('{recipe['cuisine']}');\n")
        fat_content = recipe['dietary_info'][0]
        protein_content = recipe['dietary_info'][1]
        carbs_content = recipe['dietary_info'][2]
        f2.write(f"INSERT INTO dietaryinfo (recipe_id, fat_content, protein_content, hydrocarbon_content) VALUES ({index + 1}, {fat_content}, {protein_content}, {carbs_content});\n")
        tip_1 = recipe['TIP1'].replace('\'','')
        tip_2 = recipe['TIP2'].replace('\'','')
        tip_3 = recipe['TIP3'].replace('\'','')
        prep_int = recipe['prep_time'].split(' ')[0]
        cook_int = recipe['cook_time'].split(' ')[0]
        prep_time = "00:" + prep_int + ":00"
        cook_time = "00:" + cook_int + ":00"
        f2.write(f"INSERT INTO Recipes (name, short_desc , recipe_type, difficulty_level, TIP1 ,TIP2 ,TIP3, prep_time, cook_time, cuisine_id) VALUES ('{recipe['name']}', '{recipe['short_desc']}', '{recipe['recipe_type']}', '{recipe['difficulty_level']}', '{tip_1}', '{tip_2}', '{tip_3}', '{prep_time}', '{cook_time}', {cuisines.index(recipe['cuisine']) + 1});\n")
            
```

```python
#Choose one to four random.randintom cuisines for each chef to specialize in 
cuisine_name = {}
with open("DML_SCRIPT.sql", "a", encoding="utf-8") as f2 , open("Chefs_dummy_data.json", "r") as f1:
    f2.write("\n")
    data = json.load(f1)
    chef_specializes = [[] for i in range(0, len(data))]
    for chef in data:
        already_specialized = []
        chef_id = data.index(chef) + 1
        num_cuisines = random.randint(1,4)
        for i in range(0, num_cuisines):
            cuisine_id = random.randint(1, len(cuisines))
            while cuisine_id in already_specialized:
                cuisine_id = random.randint(1, len(cuisines))
            already_specialized.append(cuisine_id)
            chef_specializes[chef_id - 1].append(cuisine_id)
            f2.write(f"INSERT INTO specialises_in (chef_id, cuisine_id) VALUES ({chef_id}, {cuisine_id});\n")
chef_does_recipe = [[] for i in range(0, len(data))]
with open("DML_SCRIPT.sql", "a", encoding="utf-8") as f2 , open("Chefs_dummy_data.json", "r") as f1 , open("Recipes.json", "r", encoding="utf-8") as f3:
    f2.write("\n")
    data = json.load(f1)
    recipes_info = json.load(f3)
    for chef in data:
        chef_id = data.index(chef) + 1
        candidate_recipes = {}
        max_recipes = 0
        for cuisine_id in chef_specializes[chef_id - 1]:
            for recipe in recipes_info:
                if recipe['cuisine'] == cuisines[cuisine_id - 1]:
                    max_recipes += 1
                    candidate_recipes[cuisine_id] = recipes_info.index(recipe) + 1
        num_recipes = random.randint(1, max_recipes)
        for i in range(0, num_recipes):
            for (cuisine_id, recipe_id) in candidate_recipes.items():
                if recipe_id in chef_does_recipe[chef_id - 1]:
                    continue
                toss_coin = random.randint(0,1)
                if len(chef_does_recipe[chef_id - 1]) == 0:
                    toss_coin = 1
                if toss_coin == 1:
                    f2.write(f"INSERT INTO chef_recipes (chef_id, recipe_id) VALUES ({chef_id}, {recipe_id});\n")
                    chef_does_recipe[chef_id - 1].append(recipe_id)
```

```python
#random episode generation cell
#Every Season has 10 episodes
#Every episode has 10 recipes,each from a different cuisine and a different chef
#Every episode has three judges
#No recipe,judge,chef,cuisine can take part in three consecutive episodes

with open("DML_SCRIPT.sql", "a", encoding="utf-8") as f:
    with open ("Chefs_dummy_data.json", "r") as f1:
        data = json.load(f1)
    f.write("\n")
    episode_id = 0
    for year in range(2018,2025):
        for episode in range(1, 11):
            episode_id += 1
            f.write(f"INSERT INTO Episodes (episode_number, year_played) VALUES ({episode}, {year});\n")
            already_used = []
            cuisines_used = []
            chosen_chefs = []
            with open("Recipes.json", "r", encoding="utf-8") as f1:
                recipes_info = json.load(f1)
                for i in range(0, 10):
                    chosen_chef = random.randint(0, len(data)-1)
                    while chosen_chef in chosen_chefs:
                        chosen_chef = random.randint(0, len(data)-1)
                    chosen_chefs.append(chosen_chef)
                    cuisine_id = random.choice(chef_specializes[chosen_chef])
                    threshold = 0
                    while cuisine_id in cuisines_used:
                        cuisine_id = random.choice(chef_specializes[chosen_chef])
                        threshold += 1
                        if threshold == 4:
                            break
                    if threshold == 4:
                        i -= 1
                        continue
                    cuisines_used.append(cuisine_id)
                    recipe_id = random.choice(chef_does_recipe[chosen_chef])
                    threshold = 0
                    while recipe_id in already_used or recipes_info[recipe_id - 1]['cuisine'] != cuisines[cuisine_id - 1]:
                        recipe_id = random.choice(chef_does_recipe[chosen_chef])
                        threshold += 1
                        if threshold == 4:
                            break
                    if threshold == 4:
                        i -= 1
                        continue
                    already_used.append(recipe_id)
                    f.write(f"INSERT INTO chefs_recipes_episode (episode_id, recipe_id, chef_id) VALUES ({episode_id}, {recipe_id}, {chosen_chef});\n")
            chefs_and_judges = chosen_chefs.copy()
            judges = []
            for _ in range(0, 3):
                judge = random.randint(1, len(data))
                while judge in chefs_and_judges:
                    judge = random.randint(1, len(data))
                judges.append(judge)
                f.write(f"INSERT INTO is_judge (judge_id, episode_id) VALUES ({judge}, {episode_id});\n")
                chefs_and_judges.append(judge)
            for chef in chosen_chefs:
                score_1 = random.randint(1, 10)
                deviation_1 = random.randint(1, 3)
                deviation_2 = random.randint(1, 3)
                score_2 = score_1 + deviation_1
                if score_2 > 10:
                    score_2 = 10
                score_3 = score_1 - deviation_2
                scores = [score_1, score_2, score_3]
                for i in range(0, 3):
                    f.write(f"INSERT INTO rates (episode_id,judge_id,contestant_id,score) VALUES ({episode_id}, {judges[i]}, {chef}, {scores[i]});\n")
            f.write("\n")
            
            
```
